```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('../data/eda_step1.csv')
submit = pd.read_csv('../data/test.csv')
len(dataset), len(submit)
```

```python
# Print the columns
# SibSp: of siblings / spouses aboard the Titanic
# Parch: of parents / children aboard the Titanic
print(dataset.columns)
```

# Understanding Outcomes by Features

```python
# Plot boxplot for Age, split by "Sex", show histogram
survived_palette = {1: 'white', 0: 'aqua'}
sns.boxplot(x='Sex', y='Age', data=dataset, hue='Survived', palette=survived_palette)
sns.swarmplot(x='Sex', y='Age', data=dataset, hue='Survived', size=4)
plt.show()
```

```python
# Plot boxplot for Age, split by "Pclass", show histogram
sns.boxplot(x='Pclass', y='Age', data=dataset,
            hue='Survived', palette=survived_palette)
sns.swarmplot(x='Pclass', y='Age', data=dataset, hue='Survived', size=4)
plt.show()
```

```python
sex_palette = {'male': 'white', 'female': 'aqua'}
sns.boxplot(x='Pclass', y='Age', data=dataset, hue='Sex', palette=sex_palette)
sns.swarmplot(x='Pclass', y='Age', data=dataset, hue='Sex', size=4)
plt.grid(True)
plt.show()
```

```python
# Plot Fare by Pclass
sns.boxplot(x='Pclass', y='Fare', data=dataset,
            hue='Survived', palette=survived_palette)
sns.swarmplot(x='Pclass', y='Fare', data=dataset, hue='Survived', size=4)
plt.grid(True)
plt.show()
```

```python
# C = Cherbourg, Q = Queenstown, S = Southampton
dataset[['Embarked']].value_counts()
```

```python
plt.figure(figsize=(10, 5))
sns.swarmplot(x='Embarked', y='Age', data=dataset, size=5, hue='Survived')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.swarmplot(x='Embarked', y='Fare', data=dataset, size=5, hue='Survived')
plt.grid(True)
plt.show()
```

```python
# Look at Siblings and Spouses
dataset[['SibSp']].value_counts()
plt.figure(figsize=(10, 5))
sns.swarmplot(x='SibSp', y='Fare', data=dataset, size=5, hue='Survived')
plt.show()
```

```python
# Look at Siblings and Spouses
dataset[['Parch']].value_counts()
plt.figure(figsize=(10, 5))
sns.swarmplot(x='Parch', y='Fare', data=dataset, size=5, hue='Survived')
plt.show()
```

```python
# Histogram of SibSp
plt.figure(figsize=(10, 5))
sns.histplot(dataset['SibSp'], bins=10)
plt.show()
```
